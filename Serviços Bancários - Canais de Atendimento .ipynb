### Previsão para os próximos dois anos - Regressão linear

```python
canais.index.name = 'trimestre'
canais = canais.sort_index()  # Garante ordem cronológica
```

```python
X = canais.index.astype(int).values.reshape(-1, 1)
y_celular = canais['celular_internet'].values
y_rede = canais['rede_bancária'].values
```

```python
modelo_celular = LinearRegression().fit(X, y_celular)
modelo_rede = LinearRegression().fit(X, y_rede)
```

```python
# Último trimestre disponível
ultimo = canais.index.astype(int).max()

# Criar os próximos trimestres
futuros = [ultimo + i for i in range(1, 21)]
X_futuro = np.array(futuros).reshape(-1, 1)

# Previsões
y_pred_celular = modelo_celular.predict(X_futuro)
y_pred_rede = modelo_rede.predict(X_futuro)

# Criar DataFrame com previsões
df_prev = pd.DataFrame({
    'celular_internet': y_pred_celular,
    'rede_bancária': y_pred_rede
}, index=futuros)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(canais.index, canais['celular_internet'], label='Celular/Internet - Real')
plt.plot(canais.index, canais['rede_bancária'], label='Rede Bancária - Real')
plt.plot(df_prev.index, df_prev['celular_internet'], '--', label='Celular/Internet - Previsto')
plt.plot(df_prev.index, df_prev['rede_bancária'], '--', label='Rede Bancária - Previsto')
plt.legend()
plt.title("Previsão de canais ao longo do tempo")
plt.xlabel("Trimestre")
plt.ylabel("Volume")
plt.grid(True)
plt.show()
```

### Previsão para os próximos dois anos - Prophet

```python
# Função para transformar o formato 20241 → 2024-01-01, etc.
def trimestre_para_data(tri):
    ano = tri // 10
    trimestre = tri % 10
    meses = {1: 1, 2: 4, 3: 7, 4: 10}
    return pd.Timestamp(year=ano, month=meses[trimestre], day=1)

# Aplicar ao índice do DataFrame
df_cel = canais[['celular_internet']].copy()
df_cel['ds'] = canais.index.astype(int).map(trimestre_para_data)
df_cel = df_cel.rename(columns={'celular_internet': 'y'})[['ds', 'y']]

modelo_cel = Prophet()
modelo_cel.fit(df_cel)

# Gerar datas trimestrais para os próximos 2 anos (8 trimestres)
future = modelo_cel.make_future_dataframe(periods=8, freq='QS')  # 'QS' = Quarterly Start

forecast = modelo_cel.predict(future)

import matplotlib.pyplot as plt
modelo_cel.plot(forecast)
plt.title("Previsão - Canal: Celular/Internet")
plt.show()
```

```python
df_rede = canais[['rede_bancária']].copy()
df_rede['ds'] = canais.index.astype(int).map(trimestre_para_data)
df_rede = df_rede.rename(columns={'rede_bancária': 'y'})[['ds', 'y']]

modelo_rede = Prophet()
modelo_rede.fit(df_rede)

future_rede = modelo_rede.make_future_dataframe(periods=8, freq='QS')
forecast_rede = modelo_rede.predict(future_rede)

modelo_rede.plot(forecast_rede)
plt.title("Previsão - Canal: Rede Bancária")
plt.show()
```

```python
#Análise das previsões

# Última data real da série
ultima_data_real = df_cel['ds'].max()

# Filtrar apenas os dados futuros
previsao_cel = forecast[forecast['ds'] > ultima_data_real][['ds', 'yhat']].copy()
previsao_rede = forecast_rede[forecast_rede['ds'] > ultima_data_real][['ds', 'yhat']].copy()

# Renomear as colunas
previsao_cel.rename(columns={'yhat': 'celular_internet'}, inplace=True)
previsao_rede.rename(columns={'yhat': 'rede_bancária'}, inplace=True)
```

```python
#Calcular crescimento percentual trimestre a trimestre
df_previsao = pd.merge(previsao_cel, previsao_rede, on='ds')
df_previsao.head()
df_previsao['crescimento_celular_%'] = df_previsao['celular_internet'].pct_change() * 100
df_previsao['crescimento_rede_%'] = df_previsao['rede_bancária'].pct_change() * 100
df_previsao
```

 ## Conclusão final

As transações realizadas por meio do celular e internet banking mostram um crescimento consistente, com variações positivas praticamente em todos os trimestres.

As transações na rede_bancária também cresce, mas de forma mais instável. Inclusive tem uma queda significativa em 2026-01-01 (-3.56%).

A tendência é de forte crescimento no uso de canais digitais (celular/internet) em relação à rede bancária tradicional.
